# Task 2: Dimensionality Reduction with PCA:

# 1.Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaled_data=scaler.fit_transform(data)
scaled_data
```

```python
from sklearn.decomposition import PCA
pca=PCA()
```

```python
data_pca=pca.fit_transform(scaled_data)
data_pca
```

# 2.Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.

```python
explained_variance=pca.explained_variance_ratio_
np.cumsum(np.round(explained_variance,decimals=4)*100)
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.title('scree plot')
plt.xlabel('Number of Components')
plt.ylabel('Cummulative Explaned variance')
plt.show()
```

# Task 4: Clustering with PCA Data:

```python
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)  # Transform the original data

# Apply K-Means clustering to the PCA-transformed data
kmeans_pca = KMeans(n_clusters=3, random_state=42)  # Same number of clusters as original
clusters_pca = kmeans_pca.fit_predict(pca_data)  # Get the cluster labels for PCA data
```

```python
import matplotlib.pyplot as plt

# Plot the PCA-transformed data with clusters
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=clusters_pca, cmap='viridis')
plt.title('Clustering Results on PCA-Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```
